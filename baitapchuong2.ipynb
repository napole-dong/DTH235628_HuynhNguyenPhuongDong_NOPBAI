## Câu 4: Python hỗ trợ những kiểu dữ liệu cơ bản nào?

Python có các kiểu dữ liệu cơ bản sau:
- Số nguyên: `int` (ví dụ: 42)
- Số thực: `float` (ví dụ: 3.14)
- Số phức: `complex` (ví dụ: 1+2j)
- Chuỗi: `str` (ví dụ: "hello")
- Boolean: `bool` (`True`/`False`)
- Danh sách: `list` (ví dụ: `[1, 2, 3]`)
- Tuple: `tuple` (ví dụ: `(1, 2)`)
- Tập hợp: `set` (ví dụ: `{1, 2}`)
- Từ điển: `dict` (ví dụ: `{"a": 1}`)
- Kiểu rỗng: `NoneType` (`None`)

Những kiểu này có thể kết hợp và chứa lẫn nhau (ví dụ: list các dict, tuple chứa list, ...).

## Câu 6: Ý nghĩa toán tử `/`, `//`, `%`, `**`, `and`, `or`, `is`

- `/`  : chia chuẩn (float division). 5 / 2 = 2.5
- `//` : chia lấy phần nguyên (floor division). 5 // 2 = 2 (với số âm làm tròn xuống)
- `%`  : modulo (phần dư). 5 % 2 = 1
- `**` : lũy thừa. 2 ** 3 = 8
- `and`, `or` : toán tử logic (ngắn mạch). `True and False` = False; `True or False` = True
- `is` : so sánh nhận dạng (identity) — kiểm tra xem hai biến có cùng đối tượng trong bộ nhớ hay không. Ví dụ `a is b` khác `a == b` (== so sánh giá trị).

Ví dụ nhanh:
```
a = 5
b = 2
print(a / b, a // b, a % b, a ** b)
print((1 == 1) and (2 > 1), (1 == 2) or (2 > 1))
```

## Câu 7: Trình bày một số cách nhập dữ liệu từ bàn phím

- `input()` : đọc chuỗi từ bàn phím. Cần chuyển kiểu nếu muốn số.
  - Ví dụ: `n = int(input("Nhập n: "))`
- Nhập nhiều giá trị trên một dòng: dùng `split()` và chuyển kiểu cho từng phần.
  - Ví dụ: `a, b = map(int, input("Nhập a b: ").split())`
- Dùng `eval()` để đánh giá biểu thức (không an toàn nếu nhận input từ nguồn không tin cậy).
- Dùng `sys.argv` hoặc `argparse` để nhận tham số dòng lệnh (thích hợp chương trình lớn).

Ví dụ:
```
# đọc hai số cách nhau bởi dấu cách
x, y = map(float, input("Nhập x y: ").split())
```

## Câu 8: Trình bày các loại lỗi khi lập trình và cách bắt lỗi trong Python

Một số lỗi phổ biến:
- `SyntaxError`: lỗi cú pháp (thường do thiếu dấu ngoặc, hai chấm, ...).
- `NameError`: tham chiếu biến chưa được định nghĩa.
- `TypeError`: thao tác giữa các kiểu không hợp lệ.
- `ValueError`: kiểu đúng nhưng giá trị không hợp lệ khi chuyển đổi (ví dụ `int('abc')`).
- `IndexError`, `KeyError`: truy cập ngoài phạm vi của list/tuple hoặc key không tồn tại trong dict.
- `ZeroDivisionError`: chia cho 0.

Cách bắt lỗi cơ bản: dùng `try` / `except` (có thể dùng `else` và `finally`). Ví dụ:

```
try:
    x = int(input("Nhập một số nguyên: "))
    print("Bạn nhập:", x)
except ValueError:
    print("Giá trị không phải số nguyên")
except Exception as e:
    print("Lỗi khác:", e)
finally:
    print("Kết thúc khối xử lý")
```

Gợi ý: luôn bắt các ngoại lệ cụ thể trước, tránh dùng `except:` tổng quát trừ khi bạn cần xử lý mọi lỗi.
